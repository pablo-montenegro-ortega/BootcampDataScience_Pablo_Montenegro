```python
import pandas as pd

ruta_archivo = r"C:\COPIA_PABLO_MAY-23-2026\PERSONAL\BIT - DATA SCIENCE\Ejercicios_Python\Informe Atenciones.csv"
df = pd.read_csv(ruta_archivo, sep=';', encoding='utf-8')

df.columns = df.columns.str.strip()
```

```python
#1.Cual es el programa con mayor consultas por año?

df["FECHA REGISTRO"] = pd.to_datetime(
    df["FECHA REGISTRO"],
    dayfirst=True)

df["FECHA REGISTRO"] = pd.to_datetime(df["FECHA REGISTRO"])

df["AÑO"] = df["FECHA REGISTRO"].dt.year

resumen = (
    df.groupby(["AÑO", "PROGRAMA"])
      .size()
      .reset_index(name="DEMANDA")
)

# Ordenamos AÑO y luego por DEMANDA
resumen = resumen.sort_values(by=["AÑO", "DEMANDA"], ascending=[True, False])

resumen = resumen.set_index(["AÑO", "PROGRAMA"])

print(resumen)
```

```python
# 2. Cual es el programa con mayor demanda por Ciudad / Municipio?

demanda = (
    df.groupby(['CIUDAD', 'PROGRAMA'])
      .size()
      .reset_index(name='Cantidad')
)

mayor_demanda = demanda.loc[
    demanda.groupby('CIUDAD')['Cantidad'].idxmax()
]

# Ordenamos por Cantidad de Solicitudes
mayor_demanda = mayor_demanda.sort_values(by='Cantidad', ascending=False)

print (mayor_demanda)





```

```python
# 3. Cual es el promedio de edad por programa?

df['EDAD'] = pd.to_numeric(df['EDAD'], errors='coerce')

promedio = (
    df.groupby('PROGRAMA')['EDAD']
      .mean()
      .reset_index(name='Edad_promedio')

)

# Ordenamos por Edad Promedio
promedio = promedio.sort_values(by='Edad_promedio', ascending=True)

print (promedio)
```
