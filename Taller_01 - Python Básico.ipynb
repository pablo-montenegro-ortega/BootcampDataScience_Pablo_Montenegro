<span style="font-size: 24px;"><span style="color: yellow;">**Estructuras de Datos (Listas, Tuplas, Diccionarios)**</span><br>

**Crea un programa para organizar información de múltiples ventas utilizando las estructuras adecuadas según la naturaleza de los datos.**<br>

* Tupla: Crea una tupla con las categorías de productos permitidas en la tienda (ej. "Electrónica", "Ropa", "Alimentos"). Se usa una tupla porque las categorías de la tienda no cambian todos los días.
* Diccionario: Crea un diccionario que represente una venta específica. Debe tener las llaves: producto, categoria, precio, cantidad y total.
* Lista: Crea una lista vacía llamada registro_ventas y añade el diccionario de la venta que acabas de crear.

```python
# 1. CREACIÓN DE TUPLA
categorias_permitidas = ("Electrónica", "Ropa", "Alimentos", "Hogar")
# 2. INICIALIZAR LA LISTA
registro_ventas = []
producto = "Audífonos Bluetooth"
categoria = "Electrónica"
precio = 45.99
cantidad = 2
if categoria in categorias_permitidas:
    venta = {
        "producto": producto,
        "categoria": categoria,
        "precio": precio,
        "cantidad": cantidad,
        "total": round(precio * cantidad, 2)
    }
    # 3. SE ADICIONA EL REGISTRO PROCESADO A LA LISTA
    registro_ventas.append(venta)
    print("¡Venta registrada con éxito!")
    print(f"Registro actual de ventas: {registro_ventas}")
else:
    print(f"Error: La categoría '{categoria}' no está permitida en la tienda.")
```

<span style="font-size: 24px;"><span style="color: yellow;">**Estructuras de Control (Condicionales y Ciclos)**</span><br>

**Haz un programa que ayude a automatizar procesos repetitivos y aplicar reglas de negocio (lógica condicional).**

<br>1- Crea un ciclo for que recorra tu lista de registro_ventas (puedes agregar un par de ventas manuales extra al código para hacer la prueba).
<br>2- Dentro del ciclo, aplica un condicional if: Si el total de la venta es mayor a $100000, aplica un descuento del 10%.
<br>3- Imprime el reporte final mostrando el producto, el total original y el total con descuento (si aplica).

```python
## ACTUALIZACIÓN DE LA LISTA
# 1. AGREGAR REGISTROS MANUALMENTE
nuevas_ventas = [
    {"producto": "Camiseta Algodón", "categoria": "Ropa", "precio": 59000.99, "cantidad": 3, "total": 177002.97},
    {"producto": "Arroz Integral Bulto", "categoria": "Alimentos", "precio": 135000.50, "cantidad": 5, "total":675002.50 },
    {"producto": "Cafetera Goteo", "categoria": "Electrónica", "precio": 100000.00, "cantidad": 1, "total": 100000.00}
]
# 2. SE ADICIONAN NUEVOS REGISTROS A LA LISTA
registro_ventas += nuevas_ventas
for revisa in registro_ventas:
    if revisa["total"] > 100000:
        descuento = (revisa["total"] * 0.10)
    else:
        descuento = 0
    producto = revisa["producto"]
    total_inicial = revisa["total"]
    total_final = total_inicial - descuento
    print(f"Producto: {producto} - Total Inicial: {total_inicial:,.2f} - Descuento: {descuento:,.2f} - Total con Descuento: {total_final:,.2f}")
```

**Crea código reutilizable y a prueba de errores de usuario siguiendo estos parámetros.**

<br>1- Crea una función llamada calcular_ingreso que reciba dos parámetros: precio y cantidad.
<br>2- Dentro de la función, utiliza un bloque try-except. Intenta multiplicar el precio por la cantidad.
<br>3- Si ocurre un error (por ejemplo, si en lugar de un número llega un texto), el except debe atraparlo, imprimir "Error: Los valores deben ser numéricos" y retornar 0.
<br>4- Prueba tu función enviando valores correctos y luego enviando un texto para ver cómo se maneja el error.

```python
## USO DE FUNCIONES Y MANEJO DE ERRORES
def calcular_ingresos(precio, cantidad):
    try:
        total = float(precio) * float(cantidad)
        print(f"✅ - Operación realizada. El resultado de {precio} X {cantidad} es: {total:,.0f}")
        return total
    except (ValueError, TypeError):
        print(f"❌ - Operación fallida  - Error: Los valores deben ser numéricos.")

## INVOCAMOS LA FUNCIÓN
resuelve = calcular_ingresos("mil", 5)
resuelve = calcular_ingresos(1000, 5)
```
