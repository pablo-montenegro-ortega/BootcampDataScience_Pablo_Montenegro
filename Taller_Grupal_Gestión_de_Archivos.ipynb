```python
import pandas as pd 
```

```python
df = pd.read_csv("Informe Atenciones.csv",sep=";")
df.info()
```

```python
#1.Cual es el programa mas usado en cada año?

df["FECHA REGISTRO"] = pd.to_datetime(
    df["FECHA REGISTRO"],
    dayfirst=True)

df["FECHA REGISTRO"] = pd.to_datetime(df["FECHA REGISTRO"])

df["AÑO"] = df["FECHA REGISTRO"].dt.year

resumen = (
    df.groupby([ "AÑO", "PROGRAMA"])
      .size())

print (resumen)
```

```python
# 3. Cual es el promedio de edad por programa?

df['EDAD'] = pd.to_numeric(df['EDAD'], errors='coerce')

promedio = (
    df.groupby('PROGRAMA')['EDAD']
      .mean()
      .reset_index(name='Edad_promedio')

)

print (promedio)
```
